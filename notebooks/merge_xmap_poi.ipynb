```python
import pandas as pd
from pathlib import Path

poi_path = Path("../data/processed/poi")

files = [
    "amusement_and_leisure_clean.parquet",
    "arts_and_culture_clean.parquet",
    "dining_and_drinking_clean.parquet",
    "landmarks_and_nature_clean.parquet",
    "personal_and_household_services_clean.parquet",
    "retail_and_shopping_clean.parquet",
    "travel_and_lodging_clean.parquet"
]

dfs = []

for f in files:
    df = pd.read_parquet(poi_path / f)
    df["category_source"] = f.replace("_clean.parquet", "")
    dfs.append(df)

print("Loaded files:", len(dfs))
```

```python
poi_master = pd.concat(dfs, ignore_index=True)

print("POI Master shape:", poi_master.shape)
poi_master.head()
```

```python
before = poi_master.shape[0]

poi_master = poi_master.drop_duplicates(subset=["poi_id"])

after = poi_master.shape[0]

print("Before dedup:", before)
print("After dedup :", after)
```

```python
out = "../data/processed/poi/poi_master.parquet"
poi_master.to_parquet(out, index=False)

print("Saved:", out)
```

```python
import pandas as pd

poi_master = pd.read_parquet("../data/processed/poi/poi_master.parquet")

print(poi_master.shape)
poi_master.head()
```

```python
poi_categories = (
    poi_master[
        ["category_level1", "category_level2", "category_level3"]
    ]
    .drop_duplicates()
    .reset_index(drop=True)
)

print(poi_categories.shape)
poi_categories.head()
```

```python
poi_categories["category_id"] = range(1, len(poi_categories) + 1)

# ترتيب الأعمدة
poi_categories = poi_categories[
    ["category_id", "category_level1", "category_level2", "category_level3"]
]

poi_categories.head()
```

```python
poi_categories.to_parquet(
    "../data/processed/poi/poi_categories.parquet",
    index=False
)
```

```python
poi_norm = poi_master.merge(
    poi_categories,
    on=["category_level1", "category_level2", "category_level3"],
    how="left"
)

poi_norm.shape
```

```python
poi_norm = poi_norm.drop(
    columns=["category_level1", "category_level2", "category_level3"]
)

poi_norm.head()
```

```python
city_aliases = pd.read_csv("../data/processed/geo/city_aliases.csv")
selected_cities = pd.read_csv("../data/final/selected_cities.csv")
```

```python
poi_norm["city_norm"] = (
    poi_norm["locality"]
    .astype(str)
    .str.strip()
    .str.lower()
)

city_aliases["alias_norm"] = (
    city_aliases["alias_norm"]
    .astype(str)
    .str.strip()
    .str.lower()
)
```

```python
poi_norm = poi_norm.merge(
    city_aliases[["alias_norm", "city_id"]],
    left_on="city_norm",
    right_on="alias_norm",
    how="left"
)

poi_norm[["locality", "city_id"]].head(10)
```

```python
poi_norm = poi_norm.drop(
    columns=["locality", "city_norm", "alias_norm"]
)
```

```python
poi_norm = poi_norm.drop_duplicates(subset=["poi_id"])
```

```python
poi_norm = poi_norm.dropna(
    subset=["city_id", "latitude", "longitude"]
)
```

```python
poi_norm.to_parquet(
    "../data/processed/poi/poi_normalized.parquet",
    index=False
)

print("POI normalization completed ✅")
```
